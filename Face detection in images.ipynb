# Test phase for "Face detection in an image"

To detect faces in an image file, I have used the "Multi-Task Cascaded Convolutional Neural Networks" MTCNN as a detector. And used Pyplot from Matplotlib to display the results.

#### First Import all the necessory tools:

```python
from mtcnn.mtcnn import MTCNN
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
```

Insert an image(having faces) into the system, along with the image directory location.

```python
filename = 'testimage.jpg' # Saves image location & name in "filename" variable.
```

Run the MTCNN detector

```python
pixels = pyplot.imread(filename)
faces = MTCNN().detect_faces(pixels) # Detects faces in the image.
```

#### Creating a function to draw a ractangle around faces in the give image:

```python
def face_ractangles(filename, result_list):
	data = pyplot.imread(filename) # load the image as data
	pyplot.imshow(data) # plot the image
	ax = pyplot.gca() # get the context for drawing boxes
	# plot each box
	for result in result_list:
		# get coordinates
		x, y, width, height = result['box']
		# create the white ractangular shape
		rect = Rectangle((x, y), width, height, fill=False, color='white')
		# draw the box
		ax.add_patch(rect)
	pyplot.show() # shows the plot
```

Call the "face_ractangles()" function to see the boxes around face in the image.

```python
face_ractangles(filename, faces)
```

#### Creating a new function to extract all faces from the image and plot them in a row:

```python
def draw_faces(filename, result_list):
	data = pyplot.imread(filename) # load the image
	# plot each face as a subplot
	for i in range(len(result_list)): 
		# get coordinates
		x1, y1, width, height = result_list[i]['box']
		x2, y2 = x1 + width, y1 + height
		# define subplot
		pyplot.subplot(1, len(result_list), i+1)
		pyplot.axis('off')
		# plot face
		pyplot.imshow(data[y1:y2, x1:x2])
	pyplot.show() # show the plot
```

Call the "draw_faces()" function to see the boxes around face in the image.

```python
draw_faces(filename, faces)
```
